# DATAFRAMES & AGGREGATING DATA

```python
import numpy as np
```

| **Column**      | **Description**                                                                                                 |
| --------------- | --------------------------------------------------------------------------------------------------------------- |
| **PassengerId** | A unique identifier for each passenger.                                                                         |
| **Survived**    | Survival status (0 = Did not survive, 1 = Survived).                                                            |
| **Pclass**      | Passenger class (1 = First class, 2 = Second class, 3 = Third class). This is a proxy for socioeconomic status. |
| **Name**        | The full name of the passenger (includes titles such as Mr., Mrs., Miss, etc.).                                 |
| **Sex**         | Passengers gender (male or female).                                                                            |
| **Age**         | Passengers age in years (can have missing values).                                                             |
| **SibSp**       | Number of siblings or spouses aboard the Titanic.                                                               |
| **Parch**       | Number of parents or children aboard the Titanic.                                                               |
| **Ticket**      | Ticket number (sometimes includes letters).                                                                     |
| **Fare**        | Amount of money the passenger paid for the ticket.                                                              |
| **Cabin**       | Cabin number (can have missing values).                                                                         |
| **Embarked**    | Port of embarkation: C = Cherbourg, Q = Queenstown, S = Southampton.                                            |
