### Agg() method

```python
titanic.head()
```

```python
def pct30(col):return col.quantile(0.3)
```

```python
titanic["Age"].agg(pct30)
```

```python
titanic["Fare"].agg(pct30)
```

```python
titanic["Age"].agg(lambda col:col.quantile(0.3))
```

```python
titanic["Age"].agg(lambda col:col.quantile(0.3))
```

### Grouping Summary

```python
titanic[titanic["Sex"]=="male"]["Age"].mean()
```

```python
titanic[titanic["Sex"]=="female"]["Age"].mean()
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
pd.DataFrame(titanic.groupby("Sex")["Age"].mean())
```

```python
titanic.groupby(["Survived","Sex"])["Age"].count()
```

```python
pd.DataFrame(titanic.groupby(["Survived","Sex"])["Age"].count())
```

```python
titanic.groupby("Sex")["Age"].agg(["count","min","max"])
```

```python
titanic.groupby(["Survived","Sex"])["Age"].agg(["count","min","max"])
```

### Pivots table

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.pivot_table(values="Age",index="Sex")
```

```python
titanic.pivot_table(values="Age",index="Sex",aggfunc="count")
```

```python
titanic.pivot_table(values="Age",index="Sex",aggfunc="mean")
```

```python
import scipy
titanic.pivot_table(values="Age",index="Sex",aggfunc=scipy.stats.mode)
```

```python
titanic.pivot_table(values="Age",index="Sex",columns = "Survived")
```

```python
titanic.pivot_table(values="Age",index="Sex",columns = "Survived",fill_value=0,margins=True)
```

```python
titanic.pivot_table(values="Age",index="Sex",columns = "Survived",fill_value=0,margins=True, margins_name="mean")
```
